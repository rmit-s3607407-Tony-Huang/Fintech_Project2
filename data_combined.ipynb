```python
import pandas as pd
```

```python
machine_learning_df = pd.read_csv('Data/machine_learning_df.csv')
```

```python
reddit_buzzwords_df = pd.read_csv('reddit_buzzwords.csv')
```

```python
google_buzzwords_df = pd.read_csv('google_buzzwords.csv')
```

```python
combined_df = pd.concat([machine_learning_df, reddit_buzzwords_df, google_buzzwords_df], axis=1)
```

```python
combined_df.head()
```

```python
combined_df.columns
```

```python
combined_df = combined_df[['date', 'class', 'volume', 'google trend','Reddit positive', 'Reddit negative','Google positive', 'Google negative','reddit buzzword score','google buzzword score']]
```

```python
combined_df.set_index('date', inplace=True)
```

```python
combined_df.head()
```

```python
combined_df.to_csv('combined_df.csv')
```
